```python
import json
import random
```

```python
with open(r'C:\Users\a.goldani\projects\mmpose-synthetic-tune-datasets\cow-coco\annotations\20kp_sample98.json') as f:
    dataset = json.load(f)
```

```python
dataset.keys()
```

```python
data_idx = list(range(len(dataset['images'])))
```

```python
random.shuffle(data_idx)
```

```python
train_set = {
    'licenses': dataset['licenses'],
    'info': dataset['info'],
    'categories': dataset['categories'],
    'images': [dataset['images'][idx] for idx in data_idx[:70]],
    'annotations': [dataset['annotations'][idx] for idx in data_idx[:70]],
}

val_set = {
    'licenses': dataset['licenses'],
    'info': dataset['info'],
    'categories': dataset['categories'],
    'images': [dataset['images'][idx] for idx in data_idx[70:85]],
    'annotations': [dataset['annotations'][idx] for idx in data_idx[70:85]],
}

test_set = {
    'licenses': dataset['licenses'],
    'info': dataset['info'],
    'categories': dataset['categories'],
    'images': [dataset['images'][idx] for idx in data_idx[85:]],
    'annotations': [dataset['annotations'][idx] for idx in data_idx[85:]],
}
```

```python
with open(r'C:\Users\a.goldani\projects\mmpose-synthetic-tune-datasets\cow-coco\annotations\20kp-sample98-train.json', 'w') as f:
    json.dump(train_set, f)
    
with open(r'C:\Users\a.goldani\projects\mmpose-synthetic-tune-datasets\cow-coco\annotations\20kp-sample98-val.json', 'w') as f:
    json.dump(val_set, f)
    
with open(r'C:\Users\a.goldani\projects\mmpose-synthetic-tune-datasets\cow-coco\annotations\20kp-sample98-test.json', 'w') as f:
    json.dump(test_set, f)
```
